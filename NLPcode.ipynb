```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatizer_tokenizer(str_input):
    words = re.sub(r"[^A-Za-z0-9\-]", " ", str_input).lower().split()
    words = [lemmatizer.lemmatize(word) for word in words]
    return words
```

```python
stop_words = stopwords.words('english')
stop_words.extend(['back','said','come','things','get','oh','one','yeah','place','would','like','know','stay','go','let','cause','doe', 'ha', 'thing', 'wa','swift','drew','la','fa','eh','em','ya','yo','na','drake','ever','ah','hey','babe','baby','right','alright'])
```

```python
from corextopic import corextopic as ct
from corextopic import vis_topic as vt
```

```python
corex_topic_model = ct.Corex(n_hidden=8, words=words)
corex_topic_model.fit(C_lyrics_martix, words=words, docs=df.vec)
```

```python
topics = corex_topic_model.get_topics()
for topic_n,topic in enumerate(topics):
    # w: word, mi: mutual information, s: sign
    topic = [(w,mi,s) if s > 0 else ('~'+w,mi,s) for w,mi,s in topic]
    # Unpack the info about the topic
    words,mis,signs = zip(*topic)    
    # Print topic
    topic_str = str(topic_n+1)+': '+', '.join(words)
    print(topic_str)
```

```python
def recommendation(lyric_matrix,song,cos_sim = cos_sim):
    Title = input('Please input a song: ')
    for s in song.Title:
        if Title == s:
            song_id = song[song.Title == Title].values[0][0]
    
            score = list(enumerate(cos_sim[song_id]))
    
            sorted_score = sorted(score,key = lambda x:x[1],reverse = True)
            sorted_score = sorted_score[1:]
    
            j = 0
            for item in sorted_score:
                song_title = song[song.song_id == item[0]][['Title','Artist']].values[0]
                print(j+1,song_title)
                j = j+1
                if j >6:
                    break
            else:
                print('Sorry, This song is not in dataset')
```
